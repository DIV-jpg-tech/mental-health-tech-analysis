```python
# -------------------------------
# DATA WRANGLING / DATA CLEANING
# -------------------------------

# 1. Handling missing values

# Filling missing values in 'self_employed' with 'No' (most common assumption)
df['self_employed'].fillna('No', inplace=True)

# Filling missing values in 'work_interfere' with 'Don't know'
df['work_interfere'].fillna("Don't know", inplace=True)


# 2. Cleaning 'Gender' column (VERY IMPORTANT)

# Standardizing gender values
df['Gender'] = df['Gender'].str.lower()

# Grouping similar values
df['Gender'] = df['Gender'].replace([
    'male', 'm', 'man', 'cis male', 'male-ish', 'maile', 'mal', 'msle', 'male (cis)', 'make'
], 'Male')

df['Gender'] = df['Gender'].replace([
    'female', 'f', 'woman', 'cis female', 'femake', 'female (cis)', 'femail'
], 'Female')

# Everything else → Others
df['Gender'] = df['Gender'].apply(lambda x: 'Others' if x not in ['Male', 'Female'] else x)


# 3. Handling Age column

# Removing unrealistic ages
df = df[(df['Age'] >= 15) & (df['Age'] <= 80)]


# 4. Filling missing values for categorical columns

categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)


# 5. Dropping 'state' column (too many missing values + not useful globally)

df = df.drop(columns=['state'])


# Final check
print("Data cleaning completed!")
print("Final shape:", df.shape)
```

## What all manipulations have you done and insights you found?

During the data wrangling process, several steps were performed to clean and prepare the dataset for analysis:

- Missing values in important columns like `self_employed` and `work_interfere` were handled using appropriate methods such as replacing them with meaningful default values or the most frequent category (mode).

- The `Gender` column contained inconsistent and unstructured entries (e.g., "male", "M", "female", etc.). These values were standardized into three categories: Male, Female, and Others to ensure consistency in analysis.

- The `Age` column was cleaned by removing unrealistic values (e.g., very low or very high ages), ensuring the data remains relevant and accurate.

- Columns with low analytical value and high missing data, such as `comments`, were removed. Additionally, the `state` column was dropped as it contained many missing values and was not useful for a global-level analysis.

- All remaining categorical missing values were handled using mode imputation to maintain data consistency.

### Key Insights:

- The dataset initially contained inconsistencies and missing values, especially in categorical variables, which could have impacted analysis results.

- The `Gender` column required significant cleaning, indicating real-world data is often messy and unstructured.

- Optional survey fields like `comments` had very low response rates, suggesting limited engagement in open-ended questions.

- After cleaning, the dataset became structured, consistent, and suitable for performing meaningful exploratory data analysis.
